#### From the above graph 0 for no heart disease; 1 for having heart disease

```python
# in our datset there are some cateogarical values into dummy features and scale all the values before training.
# create dummy columns using get_dummies method.

dataset = pd.get_dummies(df, columns = ['sex','cp','fbs','restecg','exang','slope','ca','thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

standardScaler = StandardScaler()
columns_to_scale = ['ï»¿age','trestbps','chol','thalach','oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
# create dependant variable
y = dataset['target']

# create independant variable
X = dataset.drop(['target'],axis = 1)
```

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,11):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(knn_classifier,X,y,cv = 10)
    knn_scores.append(score.mean())
```

```python
plt.plot([k for k in range(1,11)], knn_scores, color = 'red')
for i in range(1,11):
    plt.text(i,knn_scores[i-1],(i,knn_scores[i-1]))

plt.xticks([i for i in range(1,11)])
plt.xlabel('No. of Neighbors(k) ')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```
