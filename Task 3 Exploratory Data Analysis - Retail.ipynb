```python
data['State'].value_counts()
```

### Statewise Profit

```python
plt.figure(figsize = (10 , 10))
stpr = data.groupby(['State'])['Profit'].sum().nlargest(50)
stpr.plot.bar()
```

#### Here graph displays that california has the highest profit whereas Texas has least profit.

### Region-wise Profit

```python
plt.figure(figsize = (6 , 6))
plt.title('Region-wise Profits')
plt.pie(data['Region'].value_counts(),labels = data['Region'].value_counts().index,autopct = '%1.1f%%')
plt.show()
```

### Sales, Profit & Discount 

```python
plt.style.use('seaborn')
data.plot(kind='scatter',figsize = (12 , 6), x = 'Sales', y = 'Profit', c = 'Discount', s = 20, fontsize = 10, colormap = 'plasma')
plt.show()
```

#### More the Discount More the Sales but lesser the Profits.

### Profit and Loss for state. 

```python
pl = data.groupby('State')[['Sales','Profit']].sum().sort_values(by = 'Sales', ascending = False)
pl[:].plot.bar(color = ['red','blue'],figsize = (10 , 10))
plt.title('Profit and Sales across the States')
plt.xlabel('States')
plt.ylabel('Profit / Loss and Sales')
plt.show()
```
